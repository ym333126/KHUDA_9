```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
```

```python
# dataset repository path
url = "https://raw.githubusercontent.com/gaeng02/KHUDA-9th-ML/main/week2/dataset.csv"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
### answer init
answer = [0] * 5
```

## Task 1
df의 feature중 `sepal_length`에는 결측치(NaN)가 있습니다. <br>
`sepal_length` 평균을 구한 뒤 소숫점 2번째 자리에서 반올림 하세요. <br>
그 값으로 결측치를 채우세요. <br>
이때, 해당 평균 값을 `answer[1]`에 저장하세요. <br> <br>

\<hint> `round(value, n)` : value 값을 소숫점 n까지 반올림한다.

```python
## Input Box

df = pd.read_csv(url)
mean_length = 0
mean_length = round(df['sepal_length'].mean(),1)
df['sepal_length'].fillna(mean_length, inplace=True)


answer[1] = mean_length
print(mean_length)
```

## Task 2
df의 feature 중 `region`은 문자열 범주형입니다. <br>
`LabelEncoder`을 이용해 `region`을 숫자로 변환하세요. <br>
`transformed region` 값의 총합을 `answer[2]`에 저장하세요.

```python
## Input Box

sum_region = 0

encoder = LabelEncoder()
labels = encoder.fit_transform(df['region'])
df['region'] = labels
sum_region = sum(labels)


answer[2] = sum_region
print(sum_region)
```

## Task 3
df의 feature 중 숫자 feature (`sepal_length`, `sepal_width`, `petal_length`, `petal_width`)를 표준화하세요. <br>
표준화된 df의 0번째 행의 숫자 feature의 합을 `answer[3]`에 저장하세요.

```python
## Input Box

num_cols = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
scaler=StandardScaler()


df[num_cols] = scaler.fit_transform(df[num_cols])




## Don't Touch!
answer[3] = round(df[num_cols].iloc[0].sum(), 6)
print(answer[3])
```

## Task 4
Task 3에서 완료한 df에 PCA를 적용해 2개의 주성분으로 축소하세요. <br>
2개 주성분 누적 설명분산을 `answer[4]`에 저장하세요.

```python
## Input Box

pca =PCA(n_components=2)
df_pca = pca.fit_transform(df)


## Don't Touch!
answer[4] = round(pca.explained_variance_ratio_.sum(),6)
print(answer[4])
```
